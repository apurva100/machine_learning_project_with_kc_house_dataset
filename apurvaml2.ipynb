```python
x_train_1=data.iloc[:,1:]
y_train_1=data.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train_1,x_test_1,y_train_1,y_test_1=train_test_split(x_train_1,y_train_1,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train_1,y_train_1)
lin_reg.fit(x_test_1,y_test_1)
```

```python
from sklearn.tree import DecisionTreeRegressor
DT_regressor=DecisionTreeRegressor(random_state=0)
DT_regressor.fit(x_train_1,y_train_1)
DT_regressor.score(x_test_1,y_test_1)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF_regressor=RandomForestRegressor(n_estimators=10,random_state=0)
RF_regressor.fit(x_train_1,y_train_1)
RF_regressor.score(x_test_1,y_test_1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
KNR=KNeighborsRegressor(n_neighbors=2)
KNR.fit(x_train_1,y_train_1)
KNR.score(x_test_1,y_test_1)
```
